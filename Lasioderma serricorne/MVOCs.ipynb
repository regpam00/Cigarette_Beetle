```python
# Using a list comprehension to get labels for the x-axis

trials = [trial for trial, df in df3.groupby("Trial No.")]
print(trials)
len(trials)
```

```python
# Creating separate y-values for each treatment (distances)

AF_v_Ctrl = df3.loc[0:17, "Distance moved (cm)"]
Ctrl_v_Ctrl = df3.loc[18:35, "Distance moved (cm)"]
FS_v_Ctrl = df3.loc[36:53, "Distance moved (cm)"]
Normal_v_Ctrl = df3.loc[54:71, "Distance moved (cm)"]
UV_v_Ctrl = df3.loc[72:89, "Distance moved (cm)"]

print(len(AF_v_Ctrl))
print(len(Ctrl_v_Ctrl))
print(len(FS_v_Ctrl))
print(len(Normal_v_Ctrl))
print(len(UV_v_Ctrl))
```

```python
# Plot graphs for distances

x_data = trials

plt.plot(x_data, AF_v_Ctrl, label =  "AF grain vs. Ctrl")
plt.plot(x_data, Ctrl_v_Ctrl, label =  "Ctrl vs. Ctrl")
plt.plot(x_data, FS_v_Ctrl, label =  "FS grain vs. Ctrl")
plt.plot(x_data, Normal_v_Ctrl, label =  "Normal grain vs. Ctrl")
plt.plot(x_data, UV_v_Ctrl, label =  "UV sterilized grain vs. Ctrl")

plt.title("Distance L. serricorne Traveled Under Various Treatments")
plt.xlabel("Trial Number")
plt.ylabel("Distance Moved (cm)")

plt.legend()
plt.show()
```

```python
# Creating separate y-values for each treatment (velocities)

AF_v_Ctrl_2 = df3.loc[0:17, "Velocity (cm/s)"]
Ctrl_v_Ctrl_2 = df3.loc[18:35, "Velocity (cm/s)"]
FS_v_Ctrl_2 = df3.loc[36:53, "Velocity (cm/s)"]
Normal_v_Ctrl_2 = df3.loc[54:71, "Velocity (cm/s)"]
UV_v_Ctrl_2 = df3.loc[72:89, "Velocity (cm/s)"]

print(len(AF_v_Ctrl_2))
print(len(Ctrl_v_Ctrl_2))
print(len(FS_v_Ctrl_2))
print(len(Normal_v_Ctrl_2))
print(len(UV_v_Ctrl_2))
```

```python
# Plot graphs for velocities

x_data = trials

plt.plot(x_data, AF_v_Ctrl_2, label =  "AF grain vs. Ctrl")
plt.plot(x_data, Ctrl_v_Ctrl_2, label =  "Ctrl vs. Ctrl")
plt.plot(x_data, FS_v_Ctrl_2, label =  "FS grain vs. Ctrl")
plt.plot(x_data, Normal_v_Ctrl_2, label =  "Normal grain vs. Ctrl")
plt.plot(x_data, UV_v_Ctrl_2, label =  "UV sterilized grain vs. Ctrl")

plt.title("Velocity L. serricorne Traveled Under Various Treatments")
plt.xlabel("Trial Number")
plt.ylabel("Velocity (cm/s)")

plt.legend()
plt.show()
```

```python
# Mean values

grouped_df3 = df3.groupby(["Treatment"]).mean()
grouped_treatments = ["AF vs. Ctrl", "Ctrl vs. Ctrl", "FS vs. Ctrl", "Normal vs. Ctrl", "UV vs. Ctrl"]
grouped_distance_m = list(grouped_df3.loc[:, "Distance moved (cm)"])
grouped_velocity_m = list(grouped_df3.loc[:, "Velocity (cm/s)"])

print(grouped_treatments)
print(grouped_distance_m)
print(grouped_velocity_m)

#"AF grain vs. Ctrl", "Ctrl vs. Ctrl", "FS grain vs. Ctrl", "Normal Grain vs. Ctrl", "UV Sterilized Grain vs. Ctrl"
```

```python
# Standard deviation values

grouped_df3_2 = df3.groupby(["Treatment"]).std()
grouped_treatments = ["AF vs. Ctrl", "Ctrl vs. Ctrl", "FS vs. Ctrl", "Normal vs. Ctrl", "UV vs. Ctrl"]
grouped_distance_std = list(grouped_df3_2.loc[:, "Distance moved (cm)"])
grouped_velocity_std = list(grouped_df3_2.loc[:, "Velocity (cm/s)"])

print(grouped_treatments)
print(grouped_distance_std)
print(grouped_velocity_std)
```

```python
mod_df5 = df5.loc[:, ["Treatment", "# Recaptured", "# Released ", "Percent Recaptured"]]
mod_df5.head()
```

```python
order = mod_df5.sort_values("Treatment")
order.head()
```

```python
print(mod_df5.groupby(["Treatment"]).sum())
```

```python
mod_df5 = mod_df5.groupby(["Treatment"]).sum()
per_recap= list(mod_df5["Percent Recaptured"]/len(order[0:8]))
per_recap
```

```python
x = [48,49,50,51,52]
y = per_recap
z = np.array(per_recap)*10
colors = np.random.rand(5)
a = ['AF','FS','NC','PC','UV']

fig, ax = plt.subplots()
bubble = ax.scatter(x, y, s = z, c = colors)

bubble.set_edgecolor('face')
bubble.set_linewidth(2.5)

for i in range(len(x)):
    plt.text(x[i]+0.20, y[i], str(y[i])+"%", ha ='left', va ='top') # horizontal/vertical alignment

for i in range(len(x)):
    plt.text(x[i]+0.20, y[i]+0.60, str(a[i]), ha ='left', va ='center')

plt.title("Averaged Percentage of Recaptured Bugs for Each Treatment", loc = "center")
plt.axis('off')
plt.show()
```
